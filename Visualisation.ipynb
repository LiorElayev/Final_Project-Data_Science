```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("player_data_updated.csv")
df.drop('Unnamed: 0',axis='columns', inplace=True)
```

```python
df
```

```python
df['citizenship'].value_counts().head(30).plot(kind='bar', figsize=(15,8))
plt.title('Player citizenship', fontsize=20)
plt.xlabel("citizenship")
plt.ylabel('Amount')
```

```python
df['transfer_count'].value_counts().plot(kind='bar', figsize=(10,5), color='#FF0000')
plt.title('Transferring between teams', fontsize=20)
plt.xlabel("Transfer count")
plt.ylabel('Amount')
```

```python
df['foot'].value_counts()
```

```python
labels = 'Left', 'Right', 'Both'
sizes = [df['foot'].value_counts()[1], df['foot'].value_counts()[2], df['foot'].value_counts()[3]]
explode = (0.1, 0.1, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=150)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Player foot', fontsize=15)
plt.show()
```

```python
df['position'].value_counts().plot(kind='bar', figsize=(10,5), color='#00FF1E')
plt.title('Player Position', fontsize=20)
plt.xlabel("Position")
plt.ylabel('Amount')
```

```python
df['current_club'].value_counts().head(50).plot(kind='bar', figsize=(15,8))
plt.title('Player current club', fontsize=20)
plt.xlabel("current club")
plt.ylabel('Amount')
```
