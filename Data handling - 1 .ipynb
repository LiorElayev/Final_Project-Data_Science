```python
import pandas as pd
```

```python
##import player_list.csv ans player_data-original.csv files and build dataframe##
df_player_list = pd.read_csv("player_list.csv")
df_player_data = pd.read_csv("player_data-original.csv")
df_player_data.drop('Unnamed: 0',axis='columns', inplace=True)
```

```python
df_transfer_history = pd.read_csv("transfer_history.csv")
df_transfer_history.drop('Unnamed: 0',axis='columns', inplace=True)
```

```python
df_player_data
```

```python
##A loop that calculate how many times a player was transferred between teams in his career and what was his first team he played for##
first_club=list()
transfer_count=list()
for name in df_player_list['player name']:
    try:
        df_temp = df_transfer_history[df_transfer_history['player'] == name]
        first_club.append(df_temp.iloc[-1,4])
    except:
        first_club.append("unknown")
    try:
        transfer_count.append(sum(df_transfer_history['player'] == name))
    except:
        transfer_count.append(0)
```

```python
##checks lists length##
print(len(first_club))
print(len(transfer_count))
```

```python
##addition of the lists to df_player_data dataframe##
df_player_data['first_club'] = first_club
df_player_data['transfer_count'] = transfer_count
```

```python
##Export the dataframe to CSV file###
df_player_data.to_csv("player_data.csv")
```
